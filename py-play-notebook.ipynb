```python
N = 500
data = pd.DataFrame(np.random.randn(N, 2), columns=['x1', 'x2'], index=pd.date_range('2015-01-01', periods=N))
data['y'] = 2 + data.x1*5 + data.x2*(-3) + np.random.randn(N)*1.5
```

```python
data.head()
```

```python
by_year = data.groupby(lambda x: x.year)
by_year.mean().head()
```

```python
def regress(data, y, x):
    Y = data[y]
    X = data[x]
    X = sm.add_constant(X)
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
by_year.apply(regress, 'y', ['x1', 'x2'])
```

```python
results = smf.ols('y ~ x1 + x2', data).fit()
results.summary()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
fig = sm.graphics.plot_fit(results, "x1", ax=ax)
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(results, 'x2', fig=fig)
```

```python
data = pd.DataFrame(np.random.randn(50, 2), columns=['var1', 'var2'])
data['category'] = ['cat1']*10 + ['cat2']*20 + ['cat3']*20
```

```python
fig, ax = plt.subplots(figsize=(12,8))
for name, group in data.groupby('category'):
    ax.plot(group.var1, group.var2, marker='o', linestyle='', ms=10, label=name)
ax.legend(loc='best')
for label, x, y in zip(data.index, data.var1, data.var2):
    plt.annotate(label, xy=(x, y), xytext=(5, 5), textcoords='offset points')
```
